```python
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))
m = (n1 + n2)/2
print("Sua média foi {:.1f}".format(m))
if m >= 6.0:
  print("Sua média foi boa! Parabéns.")
else:
  print("Sua média foi ruim! Melhore.")
```

```python
preco = float(input("Digite o preço do produto:"))
if preco >= 100:
  desconto = preco * 0.10
  preco_final = preco - desconto
  print(f"Você ganhou desconto! Preço final: {preco_final}")
else:
  print(f"Sem desconto. Preço final: {preco}")
```

```python
nome = input("Digite seu nome:")
n1 = float(input("Digite a nota 1: "))
n2 = float(input("Digite a nota 2: "))
n3 = float(input("Digite a nota 3: "))
media = (n1 + n2 + n3) / 3
print(f"Prezado aluno: {nome}")
print(f"Média: {media:.2f}")
if media >= 9:
 print("Resultado: Parabéns, você é único e excelente, Aprovado.")
elif media >= 7:
  print("Resultado: Aprovado")
elif media >= 5:
  print("Resultado: Recuperação")
else:
  print("Resultado: REPROVADO. ESTUDE MAIS.")
```

```python
numero = int(input("Digite um número: "))
if numero >= 1 and numero <= 10:
  print("Está entre 1 e 10")
elif numero >= 11 and numero <= 20:
  print("Está entre 11 e 20")
elif numero > 20:
  print("É maior que 20")
else:
  print("É menor ou igual a 0")
```
